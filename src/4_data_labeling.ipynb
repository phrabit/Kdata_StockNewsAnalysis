```python
import pandas as pd

df = pd.read_csv('stock_price.csv')
df
```

```python
df['date'] = pd.to_datetime(df['date'])  # Convert 'date' column to datetime if it's not already

# Sort the DataFrame by stock_id and date in descending order
sorted_df = df.sort_values(by=['stock_id', 'date'], ascending=[True, False])
df = sorted_df.reset_index(drop=True)
df
```

```python
df.isnull().sum()
```

```python
from tqdm import tqdm
import datetime
for i in tqdm(range(0,len(df),65)):
  for j in range(65):
    if df.loc[i+j][1] == datetime.date(2023,4,28):
      df.loc[i+j, 'label1'] = -1
      continue
    if df.loc[i+j][3] <= df.loc[i+j][2]:
      df.loc[i+j, 'label1'] = 1
    elif df.loc[i+j][3] > df.loc[i+j][2]:
      df.loc[i+j, 'label1'] = 0
```

```python
from tqdm import tqdm_pandas, tqdm
import datetime
for i in tqdm(range(0, len(df), 65)):
    for j in range(65):
         if df.loc[i+j][1] == datetime.date(2023, 4, 28):
             df.loc[i+j, 'label2'] = -1
             continue
         if df.loc[i+j][3] < df.loc[i+j+1][2]:
             df.loc[i+j, 'label2'] = 0
         elif df.loc[i+j][3] >= df.loc[i+j+1][2]:
             df.loc[i+j, 'label2'] = 1
```

```python
df.isnull().sum()
```

```python

df['label1'] = df.label1.astype('int64')
df['label2'] = df.label2.astype('int64')
df
```

```python
df.to_csv("labeled_stock_prices.csv")
```

# 뉴스 라벨링 예제

```python
news = pd.read_csv('han_df2.csv')
news['text'][0]
```

```python
import datetime
from datetime import timedelta

for i in tqdm(range(len(news))):
    news_date_str = news.loc[i][2]  # 날짜 문자열 가져오기
    news_date = datetime.datetime.strptime(news_date_str, "%Y-%m-%d")  # 문자열을 datetime 개체로 변환 (시간 정보 제외)
    if len(df.loc[(df["stock_id"] == news.loc[i][0]) & (df["date"] == news_date.date())]) == 0:
        for j in range(1, 7):
            next_date = news_date + timedelta(days=j)
            next_date_str = next_date.strftime("%Y-%m-%d")  # 다음 날짜를 문자열 형식으로 변환
            next_date_df = df.loc[(df["stock_id"] == news.loc[i][0]) & (df["date"] == next_date_str)]
            if len(next_date_df) == 0 or next_date_df['label2'].isnull().any():
                continue
            news.loc[i, 'label'] = int(next_date_df['label2'])
        continue
    if news.loc[i][2].weekday() == 4:
        if news.loc[i][2].time() > datetime.time(18, 0):
            for j in range(1, 7):
                next_date = news_date + timedelta(days=j)
                next_date_str = next_date.strftime("%Y-%m-%d")  # 다음 날짜를 문자열 형식으로 변환
                next_date_df = df.loc[(df["stock_id"] == news.loc[i][0]) & (df["date"] == next_date_str)]
                if len(next_date_df) == 0 or next_date_df['label2'].isnull().any():
                    continue
                news.loc[i, 'label'] = int(next_date_df['label2'])
        else:
            news.loc[i, 'label'] = int(df.loc[(df["stock_id"] == news.loc[i][0]) & (df["date"] == news_date.date())]['label1'])
    else:
        if news.loc[i][2].time() > datetime.time(18):
            news.loc[i, 'label'] = int(df.loc[(df["stock_id"] == news.loc[i][0]) & (df["date"] == news_date.date())]['label2'])
        else:
            news.loc[i, 'label'] = int(df.loc[(df["stock_id"] == news.loc[i][0]) & (df["date"] == news_date.date())]['label1'])
```

```python
news.to_csv("labeled_token_news.csv")
```
