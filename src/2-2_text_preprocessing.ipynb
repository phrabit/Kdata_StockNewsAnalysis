```python
from tqdm import tqdm
import re
```

```python
import pandas as pd
data=pd.read_csv("./data/han_df2.csv", index_col=0)
data.isna().sum()
```

```python
len(data)
```

```python
data=data.drop("Unnamed: 0", axis=1)
data.head()
```

```python
data.info()
```

```python
data['token_list'] = data['stopword_tokened'].apply(lambda text: text.split(', '))
data.head()
```

```python
# 한 글자와 숫자 있는 데이터 제거!
def remove_one_word_number(tokens):
    tokens_list = [token for token in tokens if 1 < len(token)]
    pattern = re.compile(r'\d')
    tokens_of_list=[tokened for tokened in tokens_list if not re.search(pattern, tokened)]
    return tokens_of_list

tqdm.pandas()
data['list_of_tokens']=data.token_list.progress_apply(remove_one_word_number)
data
```

```python
data.isna().sum()
```

```python
df=data.drop("token_list", axis=1)
df.head()
```

```python
df=df.rename(columns={'list_of_tokens':"tokens"})
df.head()
```

```python
df=df.replace("에너지/화학","에너지_화학")
df=df.replace("철강/소재","철강_소재")
df.head()
```

```python
df.to_csv("./han_final.csv")
```

#### 진행한 파일 불러오기!

```python
import pandas as pd
df=pd.read_csv("./han_final.csv", index_col=0)
df.head()
print(df.isna().sum())
```

```python
from konlpy.tag import Hannanum
han = Hannanum() 
```

```python
df.head()
```

```python
from tqdm import tqdm
```

```python
def tokenize(text):
    tokened=han.nouns(text)  # 명사만 추출
    tokens_list = [token for token in tokened if 1 < len(token)] # 한 글자 이상인 것만 뽑아내기
    return tokens_list

tqdm.pandas()
df['token_list'] = df.tokens.progress_apply(tokenize)
df.head()
```

```python
df=df.drop("tokens", axis=1)
df.head()
```

```python
df=df.rename(columns={'token_list':"tokens"})
df.head()
```

```python
df.isna().sum()
```
