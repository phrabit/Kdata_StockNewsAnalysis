```python
import re

def clean_text(d):
  text = re.sub(r'\([^)]*\)', '', str(d))  #괄호로 둘러싸인 내용을 제거
  text = re.sub(r'\[[^]]*\]', '',str(text)) #괄호로 둘러싸인 내용을 제거
  text = re.sub(r'\<[^>]*\>', '',str(text)) # 꺽쇠 괄호로 둘러싸인 내용을 제거
  
    
  # 한글, 영문자, 숫자 및 공백 문자를 제외한 모든 문자를 공백으로 대체  
  pattern = r'[^가-힣0-9a-zA-Z\s]' 
  text = re.sub(pattern, ' ', str(text)) 
    

  text = re.sub(r'사진', ' ', str(text)) # "사진"이라는 단어를 공백으로 대체
  text = re.sub(r'.*뉴스', ' ',str(text)) # "뉴스"로 끝나는 단어 뒤의 모든 문자열을 공백으로 대체
  text = re.sub("\n", ' ', str(text)) # 줄바꿈 문자를 공백으로 대체
  text = re.sub("  +"," ", str(text)) # 두 개 이상의 공백을 하나의 공백으로 대체
  
  return text
```

```python
news['text'] = news.text.apply(clean_text)
news
```

```python
news
```

```python
okt=Okt()
use_tag=['Noun',"Verb", "Adjective", "Number","Alpha","Conjunction"]
```

```python
from tqdm import tqdm, tqdm_pandas
```

```python
def preprocess_text(text):
    pos = set(okt.pos(text))
    main_words = [str(word_pos[0]) for word_pos in pos if str(word_pos[1]) in use_tag]
    return ', '.join(main_words)
tqdm.pandas()
news['token'] = news['text'].progress_apply(preprocess_text)
```
